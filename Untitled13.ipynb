```python
jk.groupby("sex")["hours-per-week"].value_counts()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(data=jk, x='sex', hue='hours-per-week'==50, order=jk['sex'].value_counts().index, palette='Set2')
plt.title('sex vs hours per week')
plt.xlabel('sex')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### male workers are working more hours than female workers

```python
low_income=jk[jk['Income'] == ' <=50K']
```

```python
low_income
```

```python
high_income=jk[jk['Income']==" >50K"]
```

```python
high_income
```

```python
sns.countplot(x=low_income['education'],palette='pastel')
plt.xticks(rotation=90)
plt.title("education  of low income individuals") 
plt.show()
```

```python
sns.countplot(x=high_income['education'],palette='pastel')
plt.xticks(rotation=90)
plt.title("education  of high income individuals") 
plt.show()
```


### education of employees influences the income of individuals
